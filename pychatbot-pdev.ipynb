```python
import pandas as pd

# --- Conversi칩n de tipos ---
# Convertir Year-Of-Publication a num칠rico (int), manejando errores
df_books['Year-Of-Publication'] = pd.to_numeric(
    df_books['Year-Of-Publication'], errors='coerce'
).astype('Int64')  # Int64 soporta nulos

# Convertir columnas de texto repetitivo a categor칤as
df_books['Book-Author'] = df_books['Book-Author'].astype('category')
df_books['Publisher'] = df_books['Publisher'].astype('category')

# Usar ISBN como 칤ndice (clave primaria)
df_books.set_index('ISBN', inplace=True)

# --- Limpieza de NaN / null ---
# Eliminar filas con cualquier NaN en columnas cr칤ticas
df_books['Book-Author'] = df_books['Book-Author'].cat.add_categories(['Desconocido']).fillna('Desconocido')
df_books['Publisher']   = df_books['Publisher'].cat.add_categories(['Sin editorial']).fillna('Sin editorial')
df_books['Year-Of-Publication'] = df_books['Year-Of-Publication'].fillna(0000)  

#df_books = df_books.dropna(subset=['Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher'])

# Si quieres eliminar absolutamente todos los NaN en el DataFrame:
# df_books = df_books.dropna(how='any')

# --- Validaci칩n ---
print(df_books.info())
print(df_books.head())
```

```python
import streamlit as st
import pandas as pd

# --- Cargar datasets ---
df_books   = pd.read_csv("data/books.csv", encoding='ISO-8859-1', on_bad_lines='skip',sep=';')
df_ratings = pd.read_csv("data/ratings.csv", encoding='ISO-8859-1', on_bad_lines='skip',sep=';')
df_users   = pd.read_csv("data/users.csv", encoding='ISO-8859-1', on_bad_lines='skip',sep=';')

# --- Integraci칩n b치sica ---
df_books_ratings = pd.merge(df_ratings, df_books, on="ISBN", how="inner")
df_full = pd.merge(df_books_ratings, df_users, on="User-ID", how="inner")
# --- Funciones de b칰squeda ---
def get_top_books(n=10):
    return (df_full.groupby("Book-Title")["Book-Rating"]
            .mean()
            .sort_values(ascending=False)
            .head(n))

def get_books_by_author(author):
    return df_books[df_books["Book-Author"].str.contains(author, case=False, na=False)]["Book-Title"].unique()

def get_books_by_year(year):
    return df_books[df_books["Year-Of-Publication"] == year]["Book-Title"].unique()

# --- Interfaz Streamlit ---
st.title("Chatbot de Libros 游닄")

option = st.selectbox(
    "쯈u칠 quieres consultar?",
    ("Top libros mejor valorados", "Buscar por autor", "Buscar por a침o")
)

if option == "Top libros mejor valorados":
    n = st.slider("N칰mero de libros", 5, 20, 10)
    st.write(get_top_books(n))

elif option == "Buscar por autor":
    author = st.text_input("Nombre del autor:")
    if author:
        st.write(get_books_by_author(author))

elif option == "Buscar por a침o":
    year = st.number_input("A침o de publicaci칩n:", min_value=1900, max_value=2025, step=1)
    if year:
        st.write(get_books_by_year(year))
```

```python
from lightfm import LightFM
from lightfm.data import Dataset

# Crear dataset
dataset = Dataset()
dataset.fit(df_full['User-ID'], df_full['ISBN'])

# A침adir features de libros
dataset.fit_partial(items=df_full['ISBN'],
                    item_features=df_full['Book-Author'] + " " + df_full['Publisher'])

(interactions, weights) = dataset.build_interactions(
    [(row['User-ID'], row['ISBN'], row['Book-Rating']) for _, row in df_full.iterrows()]
)

item_features = dataset.build_item_features(
    [(row['ISBN'], [row['Book-Author'], row['Publisher'], str(row['Year-Of-Publication'])])
     for _, row in df_full.iterrows()]
)

# Entrenar modelo
model = LightFM(loss='warp')
model.fit(interactions, item_features=item_features, epochs=30, num_threads=4)
```
