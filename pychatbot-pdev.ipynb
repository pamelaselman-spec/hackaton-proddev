```python
import pandas as pd

# Leer el CSV
df_books = pd.read_csv('data/books.csv',encoding='ISO-8859-1', on_bad_lines='skip',sep=';')
# Mostrar las primeras filas
#print(df_books.head())
# Información general del dataset
#print(df_books.info())
# Estadísticas básicas
#print(df_books.describe(include="all"))

print(df_books.columns)

print(df_books.dtypes)
```

```python
import pandas as pd

# --- Conversión de tipos ---
# Convertir Year-Of-Publication a numérico (int), manejando errores
df_books['Year-Of-Publication'] = pd.to_numeric(
    df_books['Year-Of-Publication'], errors='coerce'
).astype('Int64')  # Int64 soporta nulos

# Convertir columnas de texto repetitivo a categorías
df_books['Book-Author'] = df_books['Book-Author'].astype('category')
df_books['Publisher'] = df_books['Publisher'].astype('category')

# Usar ISBN como índice (clave primaria)
df_books.set_index('ISBN', inplace=True)

# --- Limpieza de NaN / null ---
# Eliminar filas con cualquier NaN en columnas críticas
df_books['Book-Author'] = df_books['Book-Author'].cat.add_categories(['Desconocido']).fillna('Desconocido')
df_books['Publisher']   = df_books['Publisher'].cat.add_categories(['Sin editorial']).fillna('Sin editorial')
df_books['Year-Of-Publication'] = df_books['Year-Of-Publication'].fillna(0000)  

#df_books = df_books.dropna(subset=['Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher'])

# Si quieres eliminar absolutamente todos los NaN en el DataFrame:
# df_books = df_books.dropna(how='any')

# --- Validación ---
print(df_books.info())
print(df_books.head())
```

```python
print(df_books.columns)

print(df_books.dtypes)

df_books['Book-Author'].unique()
df_books['Publisher'].unique()
df_books['Year-Of-Publication'].unique()

# Conteo de libros por año
conteo_por_año = df_books['Year-Of-Publication'].value_counts().sort_index()

print(conteo_por_año)
```
